```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import sqlalchemy
```

```python
db_connection = sqlite3.connect('data/world_data.sqlite')
```

```python
#sqlite_master is a table with database schema
pd.read_sql(""" SELECT *
                FROM sqlite_master
                WHERE type='table';""",
           db_connection)
```

```python
# check the head
pd.read_sql("""SELECT *
               FROM Indicators
               LIMIT 3;""",
           db_connection)
```

```python
# how many rows?
pd.read_sql("""SELECT COUNT(*)
               FROM Indicators;""",
           db_connection)
```

```python
# checking for missing values in one column
pd.read_sql("""SELECT COUNT(*)
               FROM Indicators
               WHERE IndicatorName IS NULL;""",
           db_connection)
```

```python
# get th non null data
pd.read_sql("""SELECT *
            FROM Indicators
            WHERE IndicatorName IS NOT NULL
            LIMIT 3;""",
              db_connection)
```

```python
# Check if Null values are present in any column
pd.read_sql("""SELECT COUNT(*)
               FROM Indicators
               WHERE IndicatorName IS NULL
               OR IndicatorCode IS NULL
               OR CountryName IS NULL
               OR CountryCode IS NULL
               OR Year IS NULL
               OR Value IS NULL;""",
           db_connection)
```

```python
# Check if Null values are present in any column
connection = sqlalchemy.create_engine('sqlite:///data/world_data.sqlite')
pd.read_sql_table("Country",connection)
```

```python
# Read unique values from a column
pd.read_sql("""SELECT COUNT(DISTINCT IndicatorName)
                FROM Indicators;""", db_connection)
```

```python
# Selecting indicators with GDP in their name
pd.read_sql("""SELECT IndicatorName
                FROM Indicators
                WHERE IndicatorName LIKE '%GDP%';
                """, db_connection)
```

```python
pd.read_sql("""SELECT *
                FROM Indicators
                WHERE CountryName IN ('Brazil') AND Year>=2010
                AND IndicatorName LIKE '%GDP per capita (current US$)%';
                """, db_connection)
```

```python
pd.read_sql("""SELECT *
                FROM Indicators
                WHERE CountryName IN ('Brazil','China') AND Year>=2010
                AND IndicatorName LIKE '%GDP per capita (current US$)%';
                """, db_connection)
```

```python
pd.read_sql("""SELECT *
                FROM Indicators
                WHERE CountryName IN ('Brazil','China') AND Year BETWEEN 1990 AND 1999
                AND IndicatorName LIKE '%GDP per capita (current US$)%';
                """, db_connection)
```

```python
pd.read_sql(
    """
    SELECT CountryName,Year ,MAX(Value) AS MaxGDP
    FROM Indicators
    WHERE IndicatorName LIKE '%GDP per capita (current US$)%'
    AND Year = 2014
    """,
    db_connection
)
```
