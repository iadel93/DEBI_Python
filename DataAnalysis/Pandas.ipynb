```python
# drop missing values
data.dropna(inplace=False)
```

```python
data['Embarked'].isna().sum()
```

```python
data['Embarked'].fillna(0.0, inplace=True)
```

```python
data['Embarked'].isna().sum()
```

```python
# replace data
data['Sex'] = data['Sex'].replace(0,2)
data['Sex'].unique()
```

```python
# replace data
data['Sex'] = data['Sex'].replace({2:"Male",1:"Female"})
data['Sex'].unique()
# data['Sex'].dtype
```

```python
# rename columns
data.rename(columns={'Embarked':'Port'}, inplace=True)
data.head()
```

```python
# rename index
names = ["Passenger {}".format(i) for i in range(1,data.shape[0]+1)]
data.index = names
# data.rename(index={1:1000}, inplace=True)
data.head()
```

```python
# choose a specific row / rows based on a condition
data[data['Age'] > 30].rename({"Passenger 2":"Old Passenger"})
```

```python
# Apply a function to a column
data['Age'] = data['Age'].apply(lambda x: x/data['Age'].mean())
data.head()
```

```python
data['is_old'] = data['Age'].apply(lambda x: 1 if x > 1.3 else 0)
```
