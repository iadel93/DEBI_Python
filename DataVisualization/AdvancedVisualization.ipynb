```python
# Creaet a choropleth map
import pandas as pd
import folium
import json

data = pd.read_csv('data\\InfantMortalityRate.csv', encoding='latin1')
data.head()
```

```python
data_2019 = data[data['Year'] == 2019]
data_2019.head()
```

```python
data_2019.info()
```

```python
!pip install geopy
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="myapplication")
geolocator.geocode("Canada")
```

```python
# get the cordinates of the countries from its name
import geopy
from geopy.geocoders import Nominatim
data_2019 = data_2019.iloc[:10]
geolocator = Nominatim(user_agent='myapplication')
data_2019['coordinates'] = data_2019['Country'].apply(geolocator.geocode).apply(lambda x: (x.latitude, x.longitude))
data_2019.head()
```

```python
import json

# formulate the json file for choropleth map
locations = ", ".join(data_2019['coordinates'].apply(lambda x: str(x).replace('(',"[").replace(')',"]")).tolist())
locations

json_content = f'''
{{
  "type": "FeatureCollection",
  "features": [{{
      "type": "Feature",
      "properties": {{
        "name": "Afghanistan"
      }},
      "geometry": {{
        "type": "Polygon",
        "coordinates": [
            [
                {locations}
            ]
        ]
      }}
  }}]
    
}}
'''
json.loads(json_content)
with open('data\\countries.json', 'w') as f:
    f.write(json_content)
```

```python
# create a choropleth map
map = folium.Map(zoom_start=2)
world_geo = 'data\\countries.json'

# Ensure the column names match the data
data_2019 = data_2019.rename(columns={'Country': 'Country', 'Infant Mortality Rate': 'Infant Mortality Rate'})

folium.Choropleth(
    geo_data=world_geo,
    data=data_2019,
    columns=['Country', 'Infant Mortality Rate'],
    key_on='feature.properties.name',
    fill_color='YlGnBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Infant Mortality Rate'
).add_to(map)
map
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ariyoomotade/netflix-data-cleaning-analysis-and-visualization")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(path + '\\netflix1.csv')
data.head()
```

```python
new_data = data.set_index('show_id')
print(data.head())
new_data.loc['s6']
```

```python
print(data.info())
```

```python
# Extract listed_in column
data['listed_in'] = data['listed_in'].str.split(', ')
print(data.head())
```

```python
# transform the generes in the listed_in column to binary columns
# data = data.drop('listed_in', axis=1).join(data['listed_in'].str.join('|').str.get_dummies())
# data
# data.head()
data = data.explode('listed_in')
```

```python
data.info()
```

```python
import seaborn as sns

tmp = data[data['type']=="Movie"].iloc[:1000]
plt.figure(figsize=(20, 10))
sns.countplot(x=tmp['listed_in'])
plt.xticks(rotation=90)
plt.show()
```

```python
documentries = data[data['listed_in'] == 'Documentaries']
plt.pie(documentries['director'].value_counts().iloc[:10], labels=documentries['director'].value_counts().iloc[:10].index, autopct='%1.1f%%')
plt.show()
```

```python
# plot generes from listed in as per the director
vlad = data[data['director'] == 'Vlad Yudin']
plt.pie(vlad['listed_in'].value_counts().iloc[:10], labels=vlad['listed_in'].value_counts().iloc[:10].index, autopct='%1.1f%%')
plt.show()
```

```python
directors_group = data.groupby('director')['listed_in'].value_counts()
directors_group.to_frame()
# print(directors_group.index)
# get the top 10 comedies directors from the group
comedies_directors = directors_group[directors_group.index.get_level_values('listed_in') == 'Comedies'].sort_values(ascending=False).iloc[:10]
plt.pie(comedies_directors, labels=comedies_directors.index.get_level_values('director'), autopct='%1.1f%%')
plt.show()
```
