```python
x_points_1 = np.array([1,2,6,8])
y_points_1 = np.array([3,8,1,10])
x_points_2 = np.array([2,3,7,9])
y_points_2 = np.array([4,9,2,11])

plt.plot(x_points_1, y_points_1, x_points_2, y_points_2, marker = 'o')
plt.show()
```

Labels and Titles

```python
plt.plot(x_points_1, y_points_1, x_points_2, y_points_2, marker = 'o', linestyle = 'dotted')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two lines on same plot', loc='right')

plt.show()
```

```python
plt.plot(x_points_1, y_points_1, x_points_2, y_points_2, marker = 'o', linestyle = 'dotted')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two lines on same plot', loc='left')

plt.show()
```

enable the grid

```python
plt.plot(x_points_1, y_points_1, x_points_2, y_points_2, marker = 'o', linestyle = 'dotted')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two lines on same plot', loc='center')
plt.grid(True)
plt.show()
```

Subplots

```python
# subplot with 1 row and 2 columns and then add the number of the plot
plt.subplot(1, 2, 1)
plt.plot(x_points_1, y_points_1, marker = 'o', linestyle = 'dotted')
plt.title('First plot')

plt.subplot(1, 2, 2)
plt.scatter(x_points_2, y_points_2, marker = 'o')
plt.title('Second plot')

plt.suptitle('My sub-plots')
plt.show()
```

```python
# Modify the figure size
plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)
plt.plot(x_points_1, y_points_1, marker = 'o', linestyle = 'dotted')
plt.title('First plot')

plt.subplot(1, 2, 2)
plt.scatter(x_points_2, y_points_2, marker = 'o')
plt.title('Second plot')

plt.suptitle('My sub-plots')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(2, 2, 1)
plt.plot(x_points_1, y_points_1, marker = 'o', linestyle = 'dotted')
plt.title('First plot')

plt.subplot(2, 2, 2)
plt.scatter(x_points_2, y_points_2, marker = 'o')
plt.title('Second plot')

plt.subplot(2, 2, 3)
plt.plot(x_points_1, y_points_1, marker = 'o', linestyle = 'dotted')
plt.title('Third plot')

plt.suptitle('My sub-plots')
plt.show()
```

Cmap

```python
x_points = np.random.randint(100, size=(100))
x_points.sort()
# print(x_points)
y_points = 0.5 * x_points + np.random.normal(size=x_points.size)*10
colors  = np.linspace(0, 100, x_points.size)
# colors = ['red' if i > 50 else 'blue' for i in x_points]


plt.scatter(x_points,y_points, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
tmp = np.random.normal(size=x_points.size)
plt.hist(tmp, bins=20)
plt.show()
```

Sizes for points

```python
x_points = np.random.randint(100, size=(100))
x_points.sort()
# print(x_points)
y_points = 0.5 * x_points + np.random.normal(size=x_points.size)*10
colors  = np.linspace(0, 100, x_points.size)
# colors = ['red' if i > 50 else 'blue' for i in x_points]
# sizes = np.linspace(0, 100, x_points.size)
# sizes = [10 if i < 50 else 50 for i in x_points]


plt.scatter(x_points,y_points, sizes=x_points, c=y_points, cmap='viridis')
plt.colorbar()
plt.show()
```
