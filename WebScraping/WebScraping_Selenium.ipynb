```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd

browser = webdriver.Chrome()

browser.get('https://www.google.com/')
search_bar = browser.find_element(By.TAG_NAME,"textarea")
search_bar.send_keys("Amazon eg",Keys.ENTER)
browser.implicitly_wait(5)
links = browser.find_elements(By.XPATH,"//span[@dir='ltr']")
for link in links:
    print(link.text)
next = browser.find_element(By.XPATH, "//span[@class='oeN89d']/span[@dir='rtl']")
# if next:
#     next.click()
# browser.close()

data = pd.DataFrame({"Link":[link.text for link in links]})
print(data.head())
data.to_csv("amazon_links.csv")
```
