# Naieve Bayes

```python
from sklearn import datasets
from sklearn.naive_bayes import GaussianNB
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt
# Load the iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

print("Number samples: ", len(X))
print("Number features: ", len(X[0]))

# Draw a histogram of the iris dataset for each class
plt.figure()
plt.hist(X[y == 0, 0], bins=30, alpha=0.5, color='red', label='Setosa')
plt.hist(X[y == 1, 0], bins=30, alpha=0.5, color='blue', label='Versicolor')
plt.hist(X[y == 2, 0], bins=30, alpha=0.5, color='green', label='Virginica')
plt.legend(loc='upper right')
plt.title('Histogram of the iris dataset')
plt.xlabel('Sepal length (cm)')
plt.ylabel('Frequency')
plt.show()

```

```python
# Train a Gaussian Naive Bayes classifier
# gnb = GaussianNB(priors=[1.0 , 0.0, 0.0])
gnb = GaussianNB()
gnb.fit(X, y)

# Display the decision boundary
DecisionBoundaryDisplay.from_estimator(
        gnb,
        X,
        response_method="predict",
        cmap=plt.cm.Spectral,
        alpha=0.8,
        xlabel=iris.feature_names[0],
        ylabel=iris.feature_names[1],
    )

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral, edgecolors='k')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('Decision boundary of the Gaussian Naive Bayes classifier')
plt.show()

# Accuracy of the classifier
print("Accuracy: ", gnb.score(X, y))
```
